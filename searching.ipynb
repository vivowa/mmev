```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import requests
```

```python
FOLDER = './'

def build_payload(slot):
    PAYLOAD = f'https://beaconcha.in/api/v1/block/{slot}'
    return PAYLOAD

def send_payload(payload):
    try:
        response = requests.get(payload)

        raw = response.json()
        raw = raw['data']

    except:
        time.sleep(2)
        return send_payload(payload)

    return raw
```
